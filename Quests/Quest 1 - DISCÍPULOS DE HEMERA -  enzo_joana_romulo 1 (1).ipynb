### Dataset: 'Penguins'

Esse dataset apresenta informações sobre penguins:
- *species:* Indica a qual espécie o indivíduo pertence
- *island:* Indica em qual ilha o animal vive
- *bill_length_mm:* Indica o tamanho horizontal do bico do animal	em milímetros
- *bill_depth_mm:* Indica o tamanho vertical do bico do animal (altura) em milímetros
- *flipper_length_mm:* Indica o tamanho da nadadeira do animal em milímetros
- *body_mass_g:* Indica a massa corporal do animal em gramas
- *sex:* Indica o sexo do indivíduo

```python
# Carregando o dataframe disponibilizado no github
df = sns.load_dataset('penguins')
```

```python
# Deixando as colunas no formato correto
df = df.convert_dtypes()
df
```

#### Computando estatísticas descritivas

```python
df.describe(include="all")
```

Percebe-se que a espécie mais frequente é a Adelie, a ilha com maior número de pinguins é a Biscoe e o estudo indicou uma predominância masculina desse animal. Além disso, o estudo não está completo, pois faltam alguns dados, como indica a linha 'count', demonstrando que não foi possível identificar tudo sobre cada indivíduo.

#### Gráfico

O gráfico escolhido foi o 'scatterplot' para ver a distribuição das espécies comparando as dimensões do bico

```python
sns.scatterplot(data=df, x="bill_depth_mm", y="bill_length_mm", hue = 'species')
plt.title("Distribição das espécies comparando o Bico")
plt.xlabel("Profundidade do bico em mm")
plt.ylabel("Comprimento do bico em mm")
```

Percebe-se uma relação entre a espécie de pinguim analisada e as dimensões de seu bico. A espécie Adelie tem um bico menos comprido e maior, a espécie Chinstrap tem um bico mais intermediário e a espécie Gentoo tem um bico mais comprido e menor  

### Dataset: ''taxis"

```python
rf = sns.load_dataset('taxis')
rf
```

pickup: data e horário de embarque.

dropoff: data e horário de desembarque.

passengers: número de passageiros que embarcaram.

distance: distância que foi percorrida entre o embarque e o desembarque, em milhas.

fare: valor cobrado pela viagem, "passagem". 

tip: valor deixado de gorjeta.

tolls: valor de pedágio.

total: o valor total pago pela viagem. No site (https://www.nyc.gov/site/tlc/passengers/taxi-fare.page) onde foram pegas as 
informações para construir a tabela disponibilizada no github, tem uma aba "Standard Metered Fare" onde explica como os preços podem variar, por exemplo, qualquer corrida já começa com 3 dólares inicialmente.

color: cor do táxi.

payment: método de pagamento.

pickup_zone: zona de embarque, área mais específica dentro da cidade.

dropoff_zone: zona de desembarque.

pickup_borough: bairro de embarque, pode englobar várias zonas.

dropoff_borough: bairro de desembarque.

```python
rf.describe(include="all")
```

```python
from matplotlib import pyplot as plt

plt.scatter(rf["distance"].head(1000), rf['total'].head(1000))
plt.title("Valor pago por distância percorrida")
plt.xlabel("Valor pago em dólares")
plt.ylabel("Distância percorrida em milhas")
plt.show()
```

Percebemos que a maioria dos preços pagos segue uma lógica relativa a distância percorrida, ou seja, quanto maior a corrida, maior o valor cobrado. Algumas excessões são notadas, provavelmente porque uma gorjeta muito alta foi paga, o passageiro não pagou a corrida, ou algum erro foi cometido na cobrança.

```python
sns.violinplot(x="pickup_borough", y="dropoff_borough", data=rf)
plt.title('Relação entre lugar de embarque e desembarque')
plt.xlabel('Bairro de embarque')
plt.ylabel('Bairro de desembarque')
plt.show()
```

Percebe-se com o gráfico, que a maior parte das corridas se iniciavam e finalizavam no mesmo bairro, ou seja, a maioria das pessoas embarcava e desembarcava no mesmo bairro, ou nas proximidades.

### Dataset: "healthexp"

```python
dados = sns.load_dataset('healthexp')
dados
```

A coluna year se refere ao ano dos dados, é um dado quantitativo discreto.
Country se refere ao país analisado, é um dado qualitativo nominal.
Spending_USD se refere à quantidade de dólares gastos per capita no país e período, é um dado quantitativo contínuo.
Life_Expectancy se refere ao número da expectativa de vida, é um dado quantitativo discreto.

```python
dados.describe()
```

```python
dados_2020 = dados.where(dados['Year'] == 2020).dropna()

plt.plot(
    dados_2020['Country'],
    dados_2020['Life_Expectancy'],
)
plt.ylabel('Expectativa de vida', color = 'blue')
plt.xlabel("Países")

plt.twinx().scatter(
    dados_2020['Country'],
    dados_2020['Spending_USD']/1000,
    color = 'red'
)
plt.ylabel('Gasto com saúde per capita em milhares de dólares', color = 'red')
```

Podemos notar que os Estados Unidos é o país com maior gasto em súde, porém tem pior expectativa de vida, enquanto o Japão tem o menor custo per capita e maior espectativa de vida.

```python
dados_Canada = dados.where(dados['Country'] == 'Canada').dropna()
dados_Canada
```

```python
plt.plot(
    dados_Canada['Year'],
    dados_Canada['Spending_USD'],
)
plt.xlabel("Ano")
plt.ylabel("Gasto em dólares")
plt.title("Gasto com saúde per capita no Canadá")
```

Custos per capita em saúde por ano do Canadá, podemos ver que ele tem a tendência de sempre aumentar, com exceção do ano de 2018. Podemos notar também que há períodos que os custos aumentam substancialmente.
